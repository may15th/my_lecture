# 트리의 정의
한 개 이상의 노드로 이루어진 유한집합

노드 - 트리의 원소
간선(edge) - 노드를 연결하는 선, 부모 노드와 자식 노드를 연결


루트 노드(root node) - 트리의 시작 노드

형제 노드 - 같은 부모 노드의 자식 노드들

조상 노드 - 간선을 따라 루트 노드까지 이르느 경로에있는 모든 노드들

자손 노드 - 서브 트리에 있는 하위 레벨의 노드들

단말노드(리프노드) - 차수가 0인노드, 자식 노드가 없는 노드

트리는 많이 그려보고따라가봐야.
