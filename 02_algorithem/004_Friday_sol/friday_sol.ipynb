# 삼성시의 버스 노선
삼성시에 있는 5,000개의 버스 정류장은 관리의 편의를 위해 1에서 5,000까지 번호가 붙어 있다.

그리고 버스 노선은 N개가 있는데, i번째 버스 노선은 번호가 Ai이상이고,

Bi이하인 모든 정류장만을 다니는 버스 노선이다.

P개의 버스 정류장에 대해 각 정류장에 몇 개의 버스 노선이 다니는지 구하는 프로그램을 작성하라

[입력]

첫 번째 줄에 테스트 케이스의 수 T가 주어진다.

각 테스트 케이스의 첫 번째 줄에는 하나의 정수 N ( 1 ≤ N ≤ 500 )이 주어진다.

다음 N개의 줄의 i번째 줄에는 두 정수 Ai, Bi ( 1 ≤ Ai ≤ Bi ≤ 5,000 )가 공백 하나로 구분되어 주어진다.

다음 줄에는 하나의 정수 P ( 1 ≤ P ≤ 500 )가 주어진다.

다음 P개의 줄의 j번째 줄에는 하나의 정수 Cj ( 1 ≤ Cj ≤ 5,000 ) 가 주어진다.

오류
1. {정류장 : 노선 수}로 정리하는 오류


풀이전략
    정류장   1   2   3   4   5  
       P    1   2   2   1   1 

    정류장번호(idx)    0   1   2   ...         5000
    cnt
    for i: A->B
        cnt[i] += 1

공식 역량 평가에서 pycharm 사용하기 때문에. 파이참 사용할 것.

```python
#bus_stop_sol

T = int(input())
for tc in range(T):
    N = int(input())        #N = 버스 노선 수
    cnt =[0] * 5001             #1~5000번 각 정류장을 지나는 노선 수
    for _ in range(N):          # N개의 노선에 대해
        A, B = map(int, input().split())        #
        for i in range(A, B+1):
            cnt[i] += 1         #현재 노선이 i번 정류장 정착
    
    print(cnt)

P = int(input())
bus_stop = [int(input()) for _ in range(P)]
# print(bus_stop)
print(f'#{tc}', end = ' ' )
for x in bus_stop:
    print(cnt[x], end = ' ')
print()



```

```python
T = int(input())
for tc in range(T):
    #1차원 연속한 1의갯수
    N, K = map(int, input().split())
    # n*m행렬 만들기
    arr = [list(map(int, input().split())) for _ in range(N)]

    ans = 0     #당넝가 들어갈 수 있는 자리의 수
    for i in range(N): #행 우선 순회
        cnt = 0         # 연속한 빈 칸(1)의 갯수
        for j in range(N):
            #cnt로 빈 칸 갯수 세고 일치할 경우 ans를 +1해줌
            # (다음 거를 못보는 경우, 1이 계속 나오는 경우) 1을 만날때마다 cnt 1씩 증가시켜줌.
            # 0을만나거나 범위를 벗어날 경우에만  cnt,k 비교
            if arr[i][j]:#(==1True)
                cnt +=1 
            if j == N-1 or arr[i][j] == 0:
                if cnt ==K:
                    ans+=1
                cnt = 0
    print(ans)

    # 아래를 열 우선 순회로 바꾼다.
    ans = 0
    for j in range(N): # 열우선 순회
        cnt = 0
        for i in range(N):
            if arr[i][j]:
                cnt+=1
            if i == N-1 or arr[i][j] == 0:
                if cnt == K:
                    ans+=1
                cnt =0

            
```

종이 꽃가루가 들어있는 풍선이 NxM 크기의 격자판에 붙어있는데, 어떤 풍선을 터뜨리면 상하좌우의 풍선이 추가로 터진다고 한다.
다음의 경우 가운데 풍선을 터뜨리면 상하좌우의 풍선이 추가로 1개씩 터지면서 총 5개의 꽃가루가 날리게 된다.

 
NxM개의 풍선에 들어있는 종이 꽃가루 개수A가 주어지면, 한 개의 풍선을 선택해 터뜨렸을 때 날릴 수 있는 꽃가루 수 중 최대값을 출력하는 프로그램을 만드시오.
(3<=N, M<=100)

 
[입력]

첫 줄에 테스트케이스 수 T, 다음 줄부터 테스트케이스 별로 첫 줄에 N과 M, 이후 N줄에 걸쳐 M개씩 풍선에 든 종이 꽃가루 개수가 주어진다.

#과 테스트케이스 번호, 빈칸에 이어 종이 꽃가루의 최대 개수를 출력한다.


```python
di = [0,1,0,-1]
dj = [1,0,-1,0]
T = int(input())

for tc in range(1,T+1):
    N, M = map(int, input().split())
    arr = [list(map(int, input().split())) for _ in range(N)]   #N*M행렬 만드는데 arr 정의문에서 N은정해줬지만 
                                                                #M은 안정해서 의아했는데 생각해보니
                                                                #M은 내가 입력하는 수만큼 입력딘다.. 신경 쓸필요 없던것.

    max_v = 0 #합의 최솟값이 0일 것이기 때문.
    for i in range(N):
        for j in range(M):
            cnt = arr[i][j]     #터트린 풍선의 꽃가루 수
            for k in range(4):      #i,j 인접에 대해 
                ni, nj = i+di[k], j +dj[k]
                #아래 조건을 만족하면(행렬 벗어나지 않으면) i,j인접 풍선까지 더함.
                if 0<=ni<N and 0<=nj<M:
                    cnt += arr[ni][nj]
            
            if max_v < cnt: #if문의 위치 어느 for문안에 있는지 주의
                            #if문 위치 어림짐작으로 쓰는 경우 많아.
                            #i,j풍선 터뜨리고 주변 풍선까지 터졌을 때
                            #최대치인지 확인하는 것!!!
                max_v = cnt
    print(f'#[tc {max_v}]')     #모든 위치에서 확인이 끝나면

```

    종이 꽃가루가 들어있는 풍선이 M개씩 N개의 줄에 붙어있고, 어떤 풍선을 터뜨리면 안에 든 종이 꽃가루 개수만큼 상하 좌우의 풍선이 추가로 터지게 되는 게임이 있다.

예를 들어 풍선에 든 꽃가루가 1개씩일 때, 가운데 풍선을 터뜨리면 상하좌우의 풍선이 추가로 1개씩 터지면서 총 5개의 꽃가루가 날리게 된다.

 

NxM개의 풍선에 들어있는 종이 꽃가루 개수A가 주어지면, 한 개의 풍선을 선택했을 때 날릴 수 있는 꽃가루의 합 중 최대값을 출력하는 프로그램을 만드시오.

(3<=N, M<=100)

 

입력

첫 줄에 테스트케이스 수 T, 다음 줄부터 테스트케이스 별로 첫 줄에 N과 M, 이후 N줄에 걸쳐 M개씩 풍선에 든 종이 꽃가루 개수가 주어진다.

 

출력

#과 테스트케이스 번호, 빈칸에 이어 종이 꽃가루의 최대 개수를 출력한다.

```python
# 이번 문제 어려운 이유는 풍선에 든 꽃가루 갯수만큼 상하좌우 풍선 터짐.
di = [0,1,0,-1]
dj = [1,0,-1,0]
T = int(input())

for tc in range(1,T+1):
    N, M = map(int, input().split())
    arr = [list(map(int, input().split())) for _ in range(N)]   #N*M행렬 만드는데 arr 정의문에서 N은정해줬지만 
                                                                #M은 안정해서 의아했는데 생각해보니
                                                                #M은 내가 입력하는 수만큼 입력딘다.. 신경 쓸필요 없던것.

    max_v = 0 #합의 최솟값이 0일 것이기 때문.
    for i in range(N):
        for j in range(M):
            cnt = arr[i][j]     #터트린 풍선의 꽃가루 수
            for k in range(4):      #i,j 인접에 대해 
                for p in range(1, arr[i][j]+1):     #늘어나는 숫자를p라고 하고 p만큼 증가하며 되겠네.
                    ni, nj = i+di[k]*p, j+dj[k]*p
                    if 0<=ni<=N and 0<=nj<M:    #행렬 칸 벗어나지 않도록 설정.
                        cnt += arr[ni][nj]      #주변칸의 풍선 꽃가루 수
            if max_v < cnt:
                max_v = cnt
    print(f'#{tc} {max_v}')
```

```python
#ladder문제
### 항상 folder따로 받아놓고 input값 따로 받아서 불러오는 게 좋다.
### vs코드 완성 후 파이참에서 그대로 실습해보자.
import sys
sys.stdin = open('input.txt')
### 아래 나오는 visited 굉장히 많이 사용
### 원본 대신에 다양한 상황 많이 사용할 경우 다른 애들영향 안미치기 때문 이 풀이 꼭 제대로 이해 & 암기 해야 된다!!!

#1. 아래로 진행해 나갈것.
# 2. 단, 좌, 우 중 갈 수 있는 곳 마주하면 그쪽으로 먼저 갈것.
# 3. 이미 지나온 길은 다시 돌아가지 않도록 함
# 4. 마지막 행(99,n)이 되었을 떄에는 종료 되어야 함
# 5. 마지막행에서 마주한 값이 2인 경우 출발 좌표 출력


#필요인자: 시작좌표(x,y)

dx = [1,0,0] #하, 좌, 우 순서의 dx,dy    #델타탐색할 때는 
dy = [0,-1,1] #방향에 대한 정의 d가 direction이구나 ㅋㅋㅋ
            #search함수가 어차피 글로벌 참조할 테고 안바뀌는 값들이니 글로벌에 쓰기

def search(x,y):
#각 시작 좌표들마다 방문 표시를 위한 2차원 리스트 만들기
    visited = [[0]*100 for _ in range(100)]
#만약 2를 찾았을때 반환해야 할 시작좌 표 y를 기록    
    original_y = y
    # 조사 조건은 x가 99가 되기 전까지
    #방문표시
    visited[x][y] = 1
    while x != 99:
        #계속 아래로 조사하면서 진행
        #조사 순서는? 하, 좌, 우? 하, 우, 좌?
        # x+1아래, y-1 좌 이동 (x,y)가 행 열임.
        for k in range(3):
            #다음 조사 위치 좌표
            nx = x +dx[k]
            ny = y +dy[k]

            #다음 조사 예정지인 (nx,ny)위치가 안전하게 조사 가능한지
            #조건으로 확인
            #다음 조사 가능 여부 조건에 data[nx][ny] == 1로 잡을 때 발생하는 문제
            if 0 <= nx < 100 and 0 <= ny < 100 and (data[nx][ny] == 1 or data[nx][ny] ==2) and visited[nx][ny] == 0:    # if 0 <= nx < 100 and 0 <= y < 100 and (data[nx][ny] !=0): #이것도 가능
                x,y = nx,ny
                # 이전에 있던 위치를 0으로 바꿔버리면 어떨까? 무한 반복 피하기 위해서!
                # data[x][y] = 0
                #새롭게 방문한 장소 방문 표시
                visited[nx][ny] = 1

    #x가 99가 된 순간
    if data[x][y] == 2:
        return original_y
    else:
        return '실패'




#모든 출발점 찾기
for _ in range(1,11):
    tc = int(input())
    #100 * 100의 2차원 리스트
    #리스트 컴프리헨션 사용x
    # data = []
    # for i in range(100):
    #     arr = list(map(int, input().split()))
    #     print(arr)
    #     data.append(arr)    #append는 리스트든 딕셔너리든 하나를 추가해주는 메서드
    data = [list(map(int, input().split())) for _ in range(100)] #리스트 컴프리헨션은 이렇게 행렬 초기ㅗ화시킬때 사용

    #모든 출발점 찾기
    for i in range(100):
        # 항상 0번째 열에서 출발
        # 0번째 열의 i번째 행의 값이 1이면
        if data[0][i] == 1:
            # print(i)    # 모든 출발점             ###아래 result부분 수정해야 될 것 같은데 lectures보고 수정
            result = search(0, i)   #코드 실행 결과 반환 저장
        if result != '실패': #조사를 끝냈는데, result에 '실패' 말고 다른 게 있다?
            #값을 찾았으니 종료.
            break
        
    print(f'#{tc}')


```
