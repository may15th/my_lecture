## 피보나치 수열
def fibo(n):
    if n < 2:
        return n
    else:
        return fibo(n-1) + fibo(n-2)

# memoization(메모이제이션)
- 앞의 예에서 피보나치 수 구하는 함수 재귀함수로 구현한 알고리즘은 문제가 있다.
fib(7)만 해도 fib(2)가 8번 호출되는 중복호출 문제 발생.

- 메모이제이션은 프로그램 실행할 떄 이전에 계산한 값을 메모리에 저장해서 매번 다시 
계산하지 않도록 함. 전체적인 실행속도를 빠르게 하는 기술. 동적 계획법의 핵심이되는 기술.
