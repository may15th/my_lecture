백트래킹 개념
- 여러 가지 선택지들이 존재하는 상황에서 한 가지를 선택
- 선택이 이루어지면 새로운 선택지들 집합이 생성
- 이런 선택을 반복하면서 최종 상태에 도달한다.

  - 올바른 선택을 계속하면 목표 상태(goal state)에 도달한다.


## 당첨 리프 노드 찾기

- 꽝 노드가지 도달하면 최근의 선택으로 되돌아와서 다시 시작한다.
- 더 이상의 선택지가 없다면 이전의 선택지로 돌아가서 다른 선택을 한다.
- A를 선택했던 상태로 되돌아간다.
- 루트까지 돌아갔을 경우 더 이상 선택지가 없다면 찾는 답이 없다.

1. 전체 경우의 수 고려
2. 가능성 없는 경우의 수를 제거(= 가지치기 puruning)

## 백트래킹과 깊이 우선 탐색의 차이

- 어떤 노드에서 출발하는 경로가 해결책으로 이어질 것 같지 않으면 더이상 그 경롤르 ㄹ따라가지 않음으로써 시도 회수를 줄임. (pruning 가지치기)


- 시간복잡도 빅O - 최악은 DFS와 동일함.
- 깊이 우선 탐색이 모든 경로를 추적하는데 비해 백트래킹은 불필요한 경로 조기 차단,.

- 최대한 많은 경우의 수 줄일 수 있는 방법 생각해야.
